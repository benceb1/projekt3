```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv('sum2.csv', sep=";")

# features adatok
features_df = df[df.columns[4:]]
# kicseréljük -1-re ahol NaN található
features_df = features_df.apply(pd.to_numeric, errors='coerce', axis=1)
```

```python
features_df
```

```python

imputer = SimpleImputer(strategy='constant', missing_values=np.nan, fill_value=-1)  # Choose an imputation strategy
normalized_features_imputed = imputer.fit_transform(features_df)
```

```python


kmeans = KMeans(n_clusters=6, random_state=42)
clusters = kmeans.fit_predict(normalized_features_imputed)
df['Cluster'] = clusters
```
